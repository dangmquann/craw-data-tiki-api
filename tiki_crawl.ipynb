```python

import time


labels_read = []
with open('label_data.txt', 'r', encoding= 'utf-8') as f:
    lines = f.readlines()
    for line in lines:
        labels_read.append(line.replace('\n', ''))

print(labels_read)
print(sorted_dict)
for i, element in enumerate(sorted_dict):
    search_tag = element
    url_template = "https://tiki.vn/search?q=<search tag>&page=2"
    encoded_product_name = urllib.parse.quote_plus(search_tag.replace('φ', ''))
    url = url_template.replace('<search tag>', encoded_product_name)

        
    product_names = []
    req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = urllib.request.urlopen(req)
    soup = BeautifulSoup(html.read(), 'html.parser')
    script_tags = soup.find_all("script", type="application/ld+json")

   
    

    # Iterate over the <script type="application/ld+json"> tags
    for script_tag in (script_tags):

        # Extract the JSON text from the <script type="application/ld+json"> tag
        
        json_text = script_tag.text

            # Load the JSON text into a Python object
        json_object = json.loads(json_text)

            # Check if the JSON object has the "@type": "Product" property
        if json_object["@type"] == "Product":

                # Get the "href" attribute of the JSON object
            product_name = json_object["name"]

             # Add the product URL to the list
            product_names.append(product_name)

    with io.open('Nov13thTikiCrawl_2.txt', 'a', encoding='utf8') as file:
        for line in product_names:
            file.write(search_tag +'__' + line + '\n')
    with io.open('Nov13thLabels_2.txt', 'a', encoding='utf8') as file:
        file.write(search_tag + '\n')
    if i % 20 == 0:
        time.sleep(2)
    
```

```python
import re
import openpyxl
# Mở file Excel mới
wb = openpyxl.Workbook()
sheet = wb.active

# Tạo hai cột có tên là "Label" và "Product"
sheet["A1"] = "Label"
sheet["B1"] = "Product"

# Sử dụng hàm re.split() để tách dữ liệu từ file txt thành hai cột
with open("Nov13thTikiCrawl.txt", "r") as f:
    for line in f:
        labels, product = re.split("__", line)
        sheet["A2"].value = labels
        sheet["B2"].value = product

# Lưu file Excel
wb.save("13_11_tiki_crawl.xlsx")
```
